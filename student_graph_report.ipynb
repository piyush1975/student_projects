```python
std_logi_long = pd.read_csv('C:/Users/HP/OneDrive/Desktop/piyush/python_csv_files/students/collegestudents.csv')
```

```python
std_logi_long
```

```python
std_logi_long.head()
```

```python
std_logi_long = std_logi_long[['latitude','longitude']]
```

```python
std_logi_long
```

```python
std_lati_long = std_logi_long.groupby('latitude').agg({'longitude':'max'}).sort_values('longitude',ascending=False)
```

```python
std_lati_long
```

```python
plt.scatter(std_lati_long.index, std_lati_long.longitude, label='Student Data', color ='#000')
plt.legend()
plt.xlabel('latitude Data')
plt.ylabel('longitude Data')
plt.show()
```

```python
std_lati_long_mean = std_logi_long.groupby('latitude').agg({'longitude':'mean'}).sort_values('longitude',ascending=False)
```

```python
std_lati_long_mean
```

```python
std_lati_long_sum = std_logi_long.groupby('latitude').agg({'longitude':'sum'}).sort_values('longitude',ascending=False)
```

```python
std_lati_long_sum
```

```python
std_lati_long_min = std_logi_long.groupby('latitude').agg({'longitude':'min'}).sort_values('longitude',ascending=True)
```

```python
std_lati_long_min
```

```python
plt.plot(std_lati_long_mean.index,std_lati_long_mean.longitude,color='black',label='mean')
plt.plot(std_lati_long_sum.index,std_lati_long_sum.longitude,color='b',label='sum')
plt.plot(std_lati_long_min.index,std_lati_long_min.longitude,label='Min',color='r')
plt.legend()
plt.xlabel('latitude Data')
plt.ylabel('longitude Data')
plt.show()
```
