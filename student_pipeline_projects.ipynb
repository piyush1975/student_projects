```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('C:/Users/HP/OneDrive/Desktop/piyush/python_csv_files/students/collegestudents.csv')
```

```python
data.head()
```

```python
X = data[['age']].head()
y = data[['name','nationality','city','gender']].head()
```

```python
X.head()
```

```python
from sklearn.pipeline import Pipeline
```

```python
y.head()
```

```python
numeric_processer = Pipeline(steps = [
        ("imputation_age_data",SimpleImputer(missing_values=np.nan,strategy="mean")),
    ("scaler",StandardScaler())
    
])
```

```python
numeric_processer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
Categories_process = Pipeline(steps = [("imputation_cate",SimpleImputer(fill_value="missing",strategy = "cate")),
                                      ("one-hot-coder",OneHotEncoder(handle_unknown="ignore"))
                                      ])
```

```python
Categories_process
```

```python
from sklearn.compose import ColumnTransformer
```

```python
processor = ColumnTransformer([("Numerical Age Data",numeric_processer,X),
                              ("Category Data",Categories_process,y)])  
```

```python
processor
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
```

```python
final_processor = make_pipeline(processor,LogisticRegression())
```
